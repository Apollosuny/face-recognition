```python
import os
import cv2
# from utils.constants import (DATASET_BASE_DIR)
```

```python
cap = cv2.VideoCapture(0)
DATASET_BASE_DIR = "../../dataset/data/"
my_name = "Trung Tran"

save_path = os.path.join(DATASET_BASE_DIR, my_name)
os.makedirs(save_path, exist_ok=True)

num_sample = 50

i = 0

capturing = False

while cap.isOpened():
    ret, frame = cap.read()
    
    if ret:
        # Hiển thị ảnh trên cửa sổ
        cv2.imshow("Capture Photo", frame)
        
        # Chờ người dùng nhấn phím 's' để bắt đầu chụp
        key = cv2.waitKey(1)
        
        if key == ord('s'):
            capturing = True
            print("Bắt đầu chụp ảnh...")
        
        # Nếu đang ở trạng thái chụp
        if capturing:
            # Lưu ảnh vào thư mục chỉ định với kích thước 250x250
            cv2.imwrite(os.path.join(save_path, f"{my_name}_{i}.jpg"), cv2.resize(frame, (350, 350)))
            print(f"Đã lưu ảnh: {my_name}_{i:04d}.jpg")
            
            i += 1
            
            # Kiểm tra nếu đủ số lượng ảnh
            if i >= num_sample:
                print("Đã chụp đủ số lượng ảnh yêu cầu.")
                break
        
        # Nhấn phím 'q' để dừng lại bất kỳ lúc nào
        if key == ord('q'):
            print("Dừng chụp ảnh.")
            break

# Giải phóng camera và đóng các cửa sổ
cap.release()
cv2.destroyAllWindows()
```
